## **NumPy**
### 로봇에 입장에서 NumPy(Numercail Python)는 단순한 라이브러리가 아니라 **고성능 수치 연산 엔진** 이다. 로봇의 **신경망 연산** 계산을 위해 특수한 구조를 사용해서, 수만 개의 벡터 연산을 한번에 처리할 수 있어서 **real-time**으로 로봇의 실시간 움직임을 계산 할 수 있는 연산의 최적화 역할을 한다.

## **벡터**
### 위 식에 본것과 같이 행렬을 이루는 기본은 벡터로 이루어져 있다. 벡터는 **차원**과 **방향**이라는 특성이 있다. 여기서 차원이란 벡터가 가진 원소의 수이며, 방향이란 열 방향인지 행 방향인지 나타낸다.

$$
\vec{L} = \begin{bmatrix} x \\ y \\ z \end{bmatrix}
\qquad
\vec{D} = \begin{bmatrix} x & y & z \end{bmatrix}
$$

### 위 행렬에서 L은 3차원 열벡터이며, D는 3차원 행백터 이다.

## 스칼라
### 스칼라는 벡터의 방향을 바꾸지 않고 크기만 조정하는 역할은 한다. 로봇의 물리학적 관점으로 전체시스템의 조절 노브의 역할을 하여 움직임을 조정할 수 있고, 기하학적 관점에서는 방향을 유지하며 크기를 확장 하여 예측할 수 있으며, 데이터 관점에서는 다른 값들의 기준을 정하기 위해 사용된다. **Scaling(크기조절), Weight(가중치), Unit(유닛 단위)** 실전 관점으로 활용되는 중요한 요소이다.
### 실전예시로 로봇 팔 이동 거리 조절을 만들어보자(ex_2)

$$
k \vec{l} = k \begin{bmatrix} x \\ y \\ z \end{bmatrix} = \begin{bmatrix} kx \\ ky \\ kz \end{bmatrix}
$$

## 내적
### 내적은 알고리즘과 많은 연산의 기본이 되는 역할을 한다. 알고리즘적 이유에서 살펴보면 서로 다른것의 유사성을 판단하기 위해서 내적을 통해 두 방향이 얼마나 일치하는 지를 알 수 있다. 내적이 크다는 것은 비슷한 방향이라는 것으로 판단할 수 있으며 음수일 경우에는 반대의 경우라 생각 할 수 있다. 로봇에 기준으로 바라보면 건설로봇을 통해서 벽돌로 벽을 쌓으려고 할떄 이미 계산해둔 이미지 벡터의 데이터베이스와 로봇이 현장에서 계산하는 데이터베이스의 내적값을 통해 원하는 프로그래밍 결과를 도출 하는 것을 확인할 수 있다. 또한, 투영 기능을 통해 수만 개의 데이터를 가진 고차원 공간에서 필요한 핵심 방향(축)으로 투영시켜 차원을 축소시킬 수 있다. 지반위에 서있는 로봇의 다리에 가해지는 복잡한 3차원 힘중에서, 실제 지면과 작용반작용하는 수직 항력만을 판단하고 싶을떄 사용된다. 현재 **딥러닝**의 알고니즘 99%는 행렬 곱셈으로 이루어져 있으며, 방대한 내적을 얼마나 빨리 계산하여 처리하는 것이 중요한 기술싸움이다.

### x(입력 벡터_데이터)
### w(가중치 벡터_중요성을 결정하는 지능)
### x $\cdot$ w(내적_데이터와 지능이 곱해지는 과정)
### 즉 딥러닝을 통해 수천만 개의 내적을 수행해서 결과값을 도출하는 것이다.

## 직교벡터 분해
### 분해는 행렬에 숨겨진 정보를 밝혀내거나 행렬을 사용하기 쉬운 형태로 만들기도 하고 데이터를 압축하는 기능이다. 수치 처리 및 암호학에서 많이 응용된다. 로봇에서는 정밀제어 분야에 활용된다. 로봇의 팔은 여러 개의 관절로 이루어져 있는데, 특정 위치로 이동할 때 수 많은 관절 각도의 조합행렬이 생기는 상황에서 이 행렬을 분해하여, **어떤 관절을 움직이는 게 가장 효율적인지** 즉각 파악이 가능하다. 또한, 분해를 통해 필요없는 데이터를 버리고 핵심적인 주성분 데이터를 확보하여 원하는 프로그램밍을 가능하게 한다.




```python
# ex_1
import numpy as np

link1 = np.array([3.0]) 
link2 = np.array([0,4])

end_effector = link1 + link2

print(f"로봇 손 위치: {end_effector}")

# np : Numpy 라이브러리를 불러올 떄 쓴 별명
# * 라이브러리란 검증된 계산식의 집합소
# .array : 배열형태로 만드는 명령어
```

```python
# ex_2
import numpy as np

robot_direction = np.array([1, 2])
# 로봇의 속도를 3배 높이거나, 거리를 3배 늘리고 싶을떄
scalar_k = 3

scaled_move = scalar_k * robot_direction

print(f"---로봇 이동 제어 데이터---")
print(f"최종 이동경로: {scaled_move}")
```
